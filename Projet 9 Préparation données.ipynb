## Rapprochement des exports dispo_alimentaire et population <a name="merge"></a>

```python
# On merge (inner) les 2 dataframes.
etude_poulet = dispo_alimentaire.merge(population, on='Pays')

etude_poulet
```

Pas de problème avec le merge, on a bien les 172 lignes.  
On a fait le choix d'un merge en inner car on veut seulement les données qui correspondent aux 2 dataframes.  

On enlève dans les pays, la France car l'entreprise est implantée en France et que notre étude concerne l'implantation à l'international.  
De plus , la Chine, les Etats-Unis, l'Inde et le Brésil vont être supprimés car ce sont des pays qui une forte production de poulet et de population. Il n'est pas pertinent de les viser pour une implantation et en plus, leurs données ne sont pas pertinentes pour l'analyse.

```python
etude_poulet.drop(etude_poulet.loc[(etude_poulet['Pays'] == 'France') | (etude_poulet['Pays'] == 'Inde') | (etude_poulet['Pays'] == 'Brésil') | (etude_poulet['Pays'] == "États-Unis d'Amérique") | (etude_poulet['Pays'] == 'Chine, continentale')].index, inplace=True)
```

```python
etude_poulet.info()
```

Les valeurs ont bien été supprimées.  
Il y a la présence de valeurs nulles, on va utiliser la moyenne pour les compléter.

```python
# Récupération de la moyenne des exportations.
moy_export = round(etude_poulet['Exportations - Quantité'].mean(),2)

# Récupération de la moyenne des importations.
moy_import = round(etude_poulet['Importations - Quantité'].mean(),2)

# Récupération de la moyenne de la production.
moy_produc = round(etude_poulet['Production'].mean(),2)
```

```python
# Ajout des moyennes.
etude_poulet['Exportations - Quantité'] = etude_poulet['Exportations - Quantité'].fillna(moy_export)
etude_poulet['Importations - Quantité'] = etude_poulet['Importations - Quantité'].fillna(moy_import)
etude_poulet['Production'] = etude_poulet['Production'].fillna(moy_produc)
```

```python
etude_poulet.info()
```

Il n'y a plus de valeurs nulles, on peut passer à l'analyse.
